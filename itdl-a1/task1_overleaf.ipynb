```python
import pickle
import pandas as pd
```

```python
with open(f'combinations_small_10_epochs.pkl', 'rb') as f:
    combinations_small = pickle.load(f)
combinations_small2 = []
for key, value in combinations_small.items():
    lil_dict= {k: v for k, v in value.items() if k != 'history'}
    lil_dict= {key:lil_dict}
    lil_dict[key] = lil_dict[key]['test_acc']
    # replace 
    combinations_small2.append(lil_dict)
combinations_small2
```

```python
import re
df = pd.DataFrame(combinations_small2).T
# set every value of column 1+ to column 0
for i in range(1, len(df.columns)):
    df[0] = df[0].combine_first(df[i])
# drop all columns except 0
df = df.drop(df.columns[1:], axis=1)
# rename index column to index
df = df.rename_axis('index').reset_index()
df['size1'] = ''
df['size2'] = ''
df['activation_function'] = ''
df['optimizer'] = ''
df['regularizer'] = ''

# in the index column remove 'keras.src.regularizers.regularizers.'
for i in range(len(df)):
    df['index'][i] = str(df['index'][i]).replace('<keras.src.regularizers.regularizers.', '')
    # remove everything after L1 or L2
    df['index'][i] = re.sub(r'L1.*', 'L1', df['index'][i])
    df['index'][i] = re.sub(r'L2.*', 'L2', df['index'][i])
    # remove ( and )
    df['index'][i] = df['index'][i].replace('(', '')
    df['index'][i] = df['index'][i].replace(')', '')
    # split by , 
    values = df['index'][i].split(',')
    # add column size1 size 2, activation function, optimizer, regularizer and add the values to the columns
    df['size1'][i] = values[0]
    df['size2'][i] = values[1]
    df['activation_function'][i] = values[2]
    df['optimizer'][i] = values[3]
    df['regularizer'][i] = values[4]
# rename column 0 to test_acc
df = df.rename(columns={0: 'test_acc'})
# sort by test_acc
df = df.sort_values(by='test_acc', ascending=False)
# remove index column
df = df.drop('index', axis=1)
df.head()
```

```python
# get the mean of every activation function
# set test_acc to float
df['test_acc'] = df['test_acc'].astype(float)
df.groupby('size2').mean().sort_values(by='test_acc', ascending=False)
```
